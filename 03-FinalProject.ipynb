## Abstract

In this project, we explored how team-level factors influenced the share of MVP votes that are received by the quarterbacks (QB) in the National Football League (NFL). While the Most Valuable Player (MVP) award is given to an individual player, the quarterback's success does rely on the whole team around them because they operate within them. We focused on team factors such as defensive performance, offensive lineman, and strength of schedule which helps determine whether it affects the MVP vote share. To study this, we merged various datasets which included the MVP voting results, defensive stats, team standings, and the quarterback's season statistics from the years of 2000-2025. Once we cleaned and merged the data, we used graphs, permutation tests and correlations to examine the relationship between the MVP vote share and team-level variables. 

To further evaluate, results indicate that team-level factors are strongly correlated with MVP vote share than individual quarterback's personal statistics. When quarterbacks are against teams with stronger defenses and allow fewer points compared to higher points, they tend to receive higher shares of MVP votes. To see those patterns in the data, we explored that data, exploratory data analysis (EDA) by using summary statistics and graphs. In order to show that, we used correlation analysis to measure the relationship between both MVP vote share and team performance factors. Finally, we used permutation test to indicate whether the differences between top MVP candidates and the other quarterback's were statistically significant.

The permutation tests showed that team performance such as strength of schedule, defensive statistics, and points allowed showed signs of statistically significant compared to individual performance such as touchdown-to-interception ratio between the groups. This shows how significant team performance is when it comes down to MVP voting. Our results indicate that MVP voting is determined by both the quarterback's performance and also the team's success as well as the team-level factors. Voters may take into consideration the team's performance when deciding an individual quarterback.

## Hypothesis


Among the top three NFL MVP candidates who happen to be quarterbacks each year since 2000, lower offensive line pressure and sack rates (measured by percentage of sacks per play), an easier strength of schedule (measured by win percentages of opposing teams last season), and lower points allowed per game by team defenses (measured by average points allowed per game during a season) will be associated with a higher regular-season MVP vote share. This hypothesis is based on prior research suggesting that quarterback performance is influenced by team factors such as pass protection and defensive support, which can inflate traditional performance metrics. For example, a strong offensive line that minimizes sack rates on the quarterback gives the quarterback more time to make good throws, an easier strength of schedule allows the quarterback to accumulate better stats by playing against weaker defenses, and a strong team defense will strip the ball from the opposing offense to give the quarterback more play time. Since MVP voting often reflects both individual performance and team success, quarterback MVP candidates benefiting from an all around good team are expected to receive a greater share of MVP votes.

### MVP Winners by Year (Dataset 1)

The dataset from Pro-Football-Reference documents winners of the Associated Press (AP) NFL Most Valuable Player Award, given annually to the player judged most valuable in the National Football League. Each row represents a single NFL season, making the primary unit of analysis the year. The key variables include the season (Year), the player’s name, position (such as QB for quarterback or RB for running back), team, and a link to detailed voting results. Because the award is decided by media voting, the dataset captures both performance recognition and historical context rather than raw game statistics alone.

The most important metrics associated with this dataset relate to voting outcomes. Voting data typically includes total vote points, number of first-place votes, and vote share (the percentage of possible votes received). These are measured in counts (number of votes) and percentages (vote share). A higher vote share indicates a stronger consensus among voters that a player was the most valuable that season. While the main table lists only winners, the linked voting pages provide additional context about how dominant a player’s season was relative to other candidates. For example, players such as Peyton Manning and Aaron Rodgers have won multiple MVP awards, suggesting sustained high-level performance across several seasons.

There are some potential limitations to consider when analyzing this dataset. First, the MVP award is based on subjective voting rather than a purely objective formula, which introduces potential bias. Media narratives, team success, popularity, and positional importance may influence outcomes. Second, the dataset only lists winners in the main table, which can obscure close races or strong performances by runners-up unless additional voting pages are consulted. Third, historical comparisons across eras may be difficult because the NFL has evolved significantly over time. Rule changes, season length, and offensive styles have all shifted, making raw statistical comparisons imperfect. Finally, because the dataset focuses solely on award outcomes, it does not capture the full distribution of player performance across the league, limiting its usefulness for comprehensive statistical modeling without additional supplemental data.

```python
## Cleaning MVP Votes (2000-2025) and Filtering for Quarterbacks
import pandas as pd

years = range(2000, 2026)
votes_df_list = list()

for year in years:
    votes_df = pd.read_csv(f'data/00-raw/mvp_voting/mvp_voting_{year}.csv', header=1)
    votes_df['year'] = year

    if year < 2022:
        votes_df.rename(columns={'Votes':'1st Place'}, inplace=True)
    votes_df_list.append(votes_df)

all_votes = pd.concat(votes_df_list)

cols_to_drop = [
    'Att.1',
    'Yds.1',
    'TD.1',
    'Rec',
    'Yds.2',
    'TD.2',
    'Solo',
    'Sk',
    'Int.1',
    '-9999',
    'Vote Pts'
]
cols_rename = {
    '1st Place': '1st_Place_Votes'
}

# dropping and renaming columns
qb_votes = all_votes.drop(columns=cols_to_drop)
qb_votes.rename(columns = cols_rename, inplace=True)

# changing share % metric
qb_votes['Share'] = qb_votes.groupby('year')['1st_Place_Votes'].transform(lambda x: x / x.sum())

# filtering to select only Quarterbacks, resetting index
qb_votes = qb_votes[qb_votes['Pos']=='QB']
qb_votes.reset_index(drop=True, inplace=True)

# qb_votes.to_csv('data/02-processed/qb_mvp_votes.csv', index=False)
qb_votes
```

### NFL Standings and Team Stats (Dataset 2)

This dataset aggregates a wide range of team and league-level data for the 2025 National Football League season, making it valuable for analyzing team performance and trends across the league. At its core, the dataset includes standings and basic team statistics, with key metrics like Wins (W), Losses (L), Ties (T), and Winning Percentage (W-L%) that measure how each team performed over the regular season. It also shows Points For (PF) and Points Against (PA), which are the total points a team scored and allowed. These are measured in simple unit counts of points scored during games. Other derived metrics include Point Differential (PD), which is PF minus PA and gives a sense of overall team dominance, and Strength of Record (SoS) or Simple Rating System (SRS), advanced measures that adjust performance for schedule difficulty.

Beyond standings, the page links to a host of offensive and defensive statistical categories, including passing, rushing, scoring, and special teams. For example, individual leaders in passing yards help indicate which players contributed most in the air, while team totals for passing yards, rushing yards, points scored, and total yards reveal offensive profiles across the league. Passing yards are counted in yards gained through completed throws, while rushing yards count ground gains. Both are measured in units of yards gained. Points scored encompasses all scoring methods (touchdowns, field goals, extra points) and is presented in total point units, offering a snapshot of offensive effectiveness.

While the dataset is comprehensive, there are some limitations to keep in mind. First, the page aggregates a very broad set of metrics but does not always show game-level or situation-specific nuances (like performance in clutch situations or under certain conditions). This could mask variability that’s important in deeper analytics. Second, some advanced metrics rely on derived formulas or assumptions (like SRS or Strength of Schedule), which can incorporate model bias or might not reflect real competitive balance perfectly. Third, because this view is a season summary, it cannot capture mid-season changes (such as injuries, trades, or coaching changes) without drilling down into game-by-game tables. Finally, the dataset is retrospective. It reflects the end results of a season, but doesn’t directly explain why certain teams performed the way they did, especially since many contributing factors (like in-game decision making or individual player context) aren’t encoded.



```python
## Data Wrangling: Converting Raw Standings to Time-Series Format
with open('data/00-raw/season_standings_raw.csv') as f:
    lines = f.readlines()

with open('data/00-raw/season_standings.csv', 'w') as f:
    f.write('Tm,W,L,T,W-L%,PF,PA,PD,MoV,SoS,SRS,OSRS,DSRS,Year\n')

    for line in lines:
        line = line.strip()

        if line.isdigit():
            year = line
            continue

        if line.startswith('Tm'):
            continue

        if line:
            parts = line.split(',')

            if len(parts) == 12:
                parts.insert(3, '0')  # insert T=0 after L

            f.write(','.join(parts) + f',{year}\n')

standings = pd.read_csv('data/00-raw/season_standings.csv')
standings['Tm'] = standings['Tm'].str.strip('*+')

# standings.to_csv('data/02-processed/season_standings.csv', index=False)
standings
```

```python
###Permutation tests are critical to see if there is an actual difference in strength of schedule, defensive simple rating score, and points allowed by defense between quarterbacks who are in top 3 MVP candidates versus those who are not

def run_permutation_test(df, col_to_test, group_col='MVP_Status', num_permutations=10000):
    """
    Runs a permutation test to see if there's a significant difference 
    in the mean of `col_to_test` between the two groups in `group_col`.
    """
    # Drop NaNs for the specific column to avoid calculation errors
    clean_df = df.dropna(subset=[col_to_test, group_col])
    
    # Calculate observed difference in means (Top 3 MVP - Not Top 3)
    group_means = clean_df.groupby(group_col)[col_to_test].mean()
    observed_diff = group_means['Top 3 MVP (QB)'] - group_means['Not Top 3']
    
    # Array to store permuted differences
    permuted_diffs = np.zeros(num_permutations)
    
    # Extract the values to shuffle
    values = clean_df[col_to_test].values
    labels = clean_df[group_col].values
    
    # Run the permutations
    for i in range(num_permutations):
        shuffled_labels = np.random.permutation(labels)
        
        # Calculate mean for the "shuffled" Top 3 group
        top3_mask = (shuffled_labels == 'Top 3 MVP (QB)')
        mean_top3 = np.mean(values[top3_mask])
        mean_not_top3 = np.mean(values[~top3_mask])
        
        permuted_diffs[i] = mean_top3 - mean_not_top3
        
    # Calculate p-value (two-tailed)
    p_value = np.sum(np.abs(permuted_diffs) >= np.abs(observed_diff)) / num_permutations
    
    print(f"--- Permutation Test Results for {col_to_test} ---")
    print(f"Observed Difference in Means: {observed_diff:.3f}")
    print(f"P-value: {p_value:.4f}")
    if p_value < 0.001:
        print("Result is statistically significant (p < 0.05). Reject the null hypothesis.\n")
    else:
        print("Result is not statistically significant. Fail to reject the null hypothesis.\n")
        
    return observed_diff, permuted_diffs, p_value

# Run tests on your variables
obs_diff_sos, perm_diffs_sos, p_sos = run_permutation_test(dummy, 'SoS')
obs_diff_dsrs, perm_diffs_dsrs, p_dsrs = run_permutation_test(dummy, 'DSRS')
obs_diff_pa, perm_diffs_pa, p_pa = run_permutation_test(dummy, 'PA')
obs_diff_td_int, perm_diffs_td_int, p_td_int = run_permutation_test(dummy, 'TD/Int')
```

* With a p value cutoff set to 0.001, we find that only the metrics that evaluate team performance show significance
* This backs up what we saw in the graphs above: that MVP voter share is much more dependent on how well a player's team does than their own individual performance
