## Research Question

Does a decrease in offensive line pressure and sack rates (percentage of how often the quarterback is tackled behind the line of scrimmage per play), weaker strength of schedule (playing against teams with low win/loss ratios from the previous season), and increase in team defensive performance (measured by points allowed per game) correlate with a higher regular-season MVP vote share for the top three NFL MVP candidates who are quarterbacks each season since 2000?

## Background and Prior Work

IMPORTANT TERMS AND ABBREVIATIONS TO KNOW: 

1. Associated Press Most Valuable Player Award (NFL AP MVP): This is an annual award presented by the Associated Press to a player in the National Football League judged to have been the most valuable in that year's regular season (distinct from the superbowl MVP, the most valuable player in the superbowl).
2. Quarterback: A player positioned behind the ball before a play starts who directs a team's offensive play. This usually involves throwing the ball to a teammate down the field, but can also involve holding onto the ball and running it down the field if no teammates are open. 
3. Running Back: An offensive player who specializes in running the ball down the field, distinct from quarterbacks who usually throw the ball, and wide receivers who usually catch the ball. In rare cases, running backs have won the NFL AP MVP award, although this has only happened 4 times since 2000. Since this study is based on factors that influence a quarterback being selected as MVP, we will disregard running backs who won the MVP award for their year and only pay attention to the top 3 quarterbacks from that year. 
4. Offensive Line: A group of players on the quarterbacks team who shield him from the other team's defense, giving him adequate time to throw the ball down the field. A strong offensive line can greatly increase a quarterbacks performance, although we want to measure the extent to which this is true. 
5. Sack: Occurs when the defense tackles the quarterback behind the line of scrimmage before they can throw a forward pass, or forces them out of bounds behind that line. It results in a loss of yardage for the offense and is a major defensive play that shifts momentum. 
6. Touchdown: The primary scoring play in American football, worth 6 points. It occurs when a player legally gains possession of the ball in the opponent's end zone, either by carrying it across the goal line or catching a pass while in the end zone.
7. Field Goal: A 3-point scoring play, usually attempted on fourth down when the offense is close enough to the opponent's end zone but unable to score a touchdown. A kicker must place-kick or drop-kick the ball through the opponent's goalposts, passing above the crossbar and between the uprights. 
8. Downs: A down in American football is one of four permitted attempts (plays) for the offense to advance the ball at least 10 yards toward the opponent’s end zone. If the team gains 10 or more yards, they earn a new set of four downs (a "first down"). Failing to gain 10 yards in four downs results in a turnover where the other team receives the ball. 
9. Points Allowed Per Game: Points Allowed per Game (PA/G) in football is a defensive statistic representing the average number of points a team's defense surrenders to opponents per game over a specific period, calculated as Total Points Allowed / Games Played. It measures defensive efficiency, with lower averages indicating a stronger defense. 

The NFL Associated Press Most Valuable Player Award is the most prestigious individual honor in professional American football, recognizing the player who provided the highest value to their team's success during the regular season. Voted on by a 50-person media panel, it highlights exceptional performance, usually for a top-performing quarterback on a winning team, and serves as a major accolade during the annual NFL Honors ceremony. However, this award is subject to a lot of controversy as many raise questions if the performance of the team surrounding MVP candidates is indirectly more valuable in the selection process than the skill of the player selected.

The 2025 NFL AP MVP race between Matthew Stafford (QB Rams) and Drake Maye (QB Patriots) already has many individuals racing to find different variables that affect MVP selection. One of these variables is the strength of schedule, a statistic that is determined by summing the win-loss records (or winning percentages) of all of a team's opponents from the previous season. Patriots have the 3rd easiest strength of schedule, while the Rams have tied for the 18th strength of schedule. While it's worth noting that Stafford did in fact receive the most recent MVP award, previous MVP winners have faced very easy teams such as four time MVP Aaron Rodgers with the Green Bay Packers.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1)

Another factor that is critical to the performance of previous MVP winners is the strength of the offensive line. The offensive line is composed of five players whose responsibility is to always protect the quarterback from defensive pressures and sacks. These players are: center, left guard, left tackle, right guard, and right tackle. A successful offensive line ensures that the pocket does not falter, preventing pressures and sacks that would hurt offensive progress made by either quarterback or running back.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2)

An overlooked statistic in MVP odds is defensive performance.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) On the surface, the defense of a team might not account for much in the way of offense. However, a good defense can stop a team short and force a punt that places the offense much closer to the endzone than expected. As a result, the defense is a critical component of the success of an offense, and by extension the MVP candidate. A strong defense is able to turn the ball over to the offense, giving more opportunities for the offense to score with elaborate plays.

## Hypothesis


Among the top three NFL MVP candidates who happen to be quarterbacks each year since 2000, lower offensive line pressure and sack rates (measured by percentage of sacks per play), an easier strength of schedule (measured by win percentages of opposing teams last season), and lower points allowed per game by team defenses (measured by average points allowed per game during a season) will be associated with a higher regular-season MVP vote share. This hypothesis is based on prior research suggesting that quarterback performance is influenced by team factors such as pass protection and defensive support, which can inflate traditional performance metrics. For example, a strong offensive line that minimizes sack rates on the quarterback gives the quarterback more time to make good throws, an easier strength of schedule allows the quarterback to accumulate better stats by playing against weaker defenses, and a strong team defense will strip the ball from the opposing offense to give the quarterback more play time. Since MVP voting often reflects both individual performance and team success, quarterback MVP candidates benefiting from an all around good team are expected to receive a greater share of MVP votes.
